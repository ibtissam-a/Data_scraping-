```python
# Import libs 

import pandas as pd
import urllib.robotparser  # we will use it to check the access to the web site and if we can scrape it or not 
from bs4 import BeautifulSoup as bs
from urllib.request import urlopen
```

```python
# We will use the robotparser to check if the web site can be scraped or not

def can_fetch(url,user_agent='*'): 
    rp = urllib.robotparser.RobotFileParser() # Create a robot parser
    rp.set_url(url + "/robots.txt")   # Add the "/robots.txt" by the end of the basic url
    rp.read()
    return rp.can_fetch(url,user_agent)


page_url = "https://wuzzuf.net/search/jobs/?q=data%20%20analyst&a=hpb"  # This is the url that we will work with 

if can_fetch(page_url):
    print(f"This url : {page_url} can be scraped.")
else:
    print(f"This url : {page_url} can't be scraped.")
```

```python
# Create the csv file where we will write our scraped data

file = open("/kaggle/working/scrap_data.csv",'w')
header = "job_title, company_name, job_adress, job_type, exp_years, exp_level \n"
file.write(header)
```

```python
# Start the scraping process

for i in range(27):    # loop over all pages
    base_url = "https://wuzzuf.net/search/jobs/?a=hpb&q=data%20analyst%20&start="+str(i)
    print(base_url)
    req = urlopen(base_url)            # Create a request 
    html = req.read()                  # Read the page html content
    req.close                          # close the request 
    soup = bs(html, "html.parser")     # Create the html parser using beautifulsoup
    blocs = soup.findAll("div", {"class":"css-1gatmva e1v1l3u10"})   # Find blocs from where we will extract informations
    
    for i in range(len(blocs)):
        
        # Scrap the job  title
        job_title = blocs[i].find("a", {"class": "css-o171kl"}).text.strip()
        job_title = job_title.replace(",","_")

        #Scrap the  company namme
        company_name = blocs[i].find("a",{"class":"css-17s97q8"}).text.strip()
        company_name = company_name.replace(",","_")
        
        #Scrap the job adress
        job_adress = blocs[i].find("span", {"class":"css-5wys0k"}).text.strip(" ")
        job_adress = job_adress.replace(", ","_")
        
        #Scrap the job type 
        job_type = blocs[i].find("div", {"class":"css-1lh32fc"}).text.strip()

        # scrap the nb years of experience
        try: 
            exp_years = blocs[i].findAll("span")
            exp_years = exp_years[3].text.strip(".")
        except IndexError:
            exp_years = "not specified"
       
        # scrap the experience level
        exp_level = blocs[i].findAll("a", {"class":"css-o171kl"})
        exp_level = exp_level[1].text
        
        # Write the scraped data 
        file.write(job_title + "," + company_name + "," + job_adress + "," + job_type + "," + exp_years + "," + exp_level + "\n")
file.close()  # We must close the file 
```
